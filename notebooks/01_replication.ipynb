# Cleaning

```python
PATH = '../data/raw'

# 3 factors CSV
fact = pd.read_csv(f'{PATH}/F-F_Research_Data_Factors.csv',
                   skiprows=3)
# Portfolios CSV
port = ff_monthly_loader(f'{PATH}/25_Portfolios_5x5.csv', skiprows=15)
fact = ff_monthly_loader(f'{PATH}/F-F_research_Data_Factors.csv', skiprows=3)

# Robustness checks
print(fact.head(), '\n')
print(fact.isna().sum())

# Robustness checks
print(fact.head(), '\n')
print(fact.isna().sum())
```

# Filtering Dates
In the paper they used the date range = [July 1963 - December 1991]

```python
# Resetting index to filter
fact.reset_index(inplace=True)
port.reset_index(inplace=True)

# Filtering date to year range
fact = fact[fact['Date'].between('1963-07-01', '1991-12-31')]
port = port[port['Date'].between('1963-07-01', '1991-12-31')]

# Robustness check
assert len(fact) == len(port), 'Observation sizes do not match'
assert fact['Date'].equals(port['Date']), "Dates don't match"

fact = fact.set_index('Date').sort_index()
port = port.set_index('Date').sort_index()
```

# Data Exploration

```python
print(fact.info())
print(port.info())

# Creating list from columns
fact_cols = fact.columns.to_list()
port_cols = port.columns.to_list()

# Applying numeric transformation
fact[fact_cols] = fact[fact_cols].apply(pd.to_numeric, errors='coerce')
port[port_cols] = port[port_cols].apply(pd.to_numeric, errors='coerce')
```
