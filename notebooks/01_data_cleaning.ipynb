# Data Source
All **replication** data will be taken from Kenneth French's Data library.

#### 1) Fama/French 5 Factors (Monthly) CSV
*We are using the historical archive 2015 version of the data, since the methodology changed afterwards*

#### 2) 25 Portfolios Formed on Size and Book-to-Market (5 x 5) CSV
Portfolios formed from sort of stocks into five **size** quintiles and five **B/M** quintiles

#### 3) 25 Portfolios Formed on Size and Operating Profitability (5 x 5) CSV
Portfolios from sort of stocks into five **size** quintiles and five **profitability** quintiles

#### 4) 25 Portfolios Formed on Size and Investment (5 x 5)  CSV
Portfolios from sort of stocks into five **size** quintiles and five **investment** quintiles

#### 5) 32 Portfolios Formed on Size, Book-to-Market, and Operating Profitability (2 x 4 x 4)  CSV
Portfolios from sort of stocks into two **size** groups and independently into four **B/M** and four **profitability** groups

#### 6) 32 Portfolios Formed on Size, Book-to-Market, and Investment (2 x 4 x 4) CSV
Portfolios from sort of stocks into two **size** groups and independently into four **B/M** and four **investment** groups


```python
import pandas as pd
import numpy as np
import fastparquet

from utils.ff_functions import ff_monthly_loader
```

# Cleaning

```python
PATH = '../data/raw'

# Loading Portfolio CSVs
port_25_me_beme = ff_monthly_loader(f'{PATH}/25_Portfolios_5x5.csv', skiprows=15)  # 25 portfolios size bm
port_25_me_op = ff_monthly_loader(f'{PATH}/25_Portfolios_ME_OP_5x5.csv', skiprows=22) # 25 portfolios ME/OP
port_25_me_inv = ff_monthly_loader(f'{PATH}/25_Portfolios_ME_INV_5x5.csv', skiprows=16) # 25 portfolios ME/INV

# Loading factor csv
fact_ff5 = ff_monthly_loader(f'{PATH}/F-F_research_Data_5_Factors_2x3.csv', skiprows=3) # 5 factor
```

```python
# Saving cleaned dataframes
fact_ff5.to_parquet(
    '../data/processed/ff5_factors_monthly.parquet',
    index=False
)

# 25 ME/SIZE
port_25_me_beme.to_parquet(
    '../data/processed/ff_portfolios_25_ME_BEME.parquet',
    index=False
)

# 25 ME/OP
port_25_me_op.to_parquet(
    '../data/processed/ff_portfolios_25_ME_OP.parquet',
    index=False
)

# 25 ME/INV
port_25_me_inv.to_parquet(
    '../data/processed/ff_portfolios_25_ME_INV.parquet',
    index=False
)
```
