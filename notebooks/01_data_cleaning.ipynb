```python
import pandas as pd
import numpy as np
import fastparquet

from utils.ff_functions import ff_monthly_loader
```

# Cleaning

```python
PATH = '../data/raw'

# Loading Portfolio CSVs
port_25_me_size = ff_monthly_loader(f'{PATH}/25_Portfolios_5x5.csv', skiprows=15)  # 25 portfolios size bm
port_25_me_op = ff_monthly_loader(f'{PATH}/25_Portfolios_ME_OP_5x5.csv', skiprows=22) # 25 portfolios ME/OP
port_25_me_inv = ff_monthly_loader(f'{PATH}/25_Portfolios_ME_INV_5x5.csv', skiprows=16) # 25 portfolios ME/INV
port_32_me_beme_inv = ff_monthly_loader(f'{PATH}/32_Portfolios_ME_BEME_INV_2x4x4_Wout_Div.csv', skiprows=16) # 32 portfolios ME/BEME/INV
port_32_me_beme_op = ff_monthly_loader(f'{PATH}/32_Portfolios_ME_BEME_OP_2x4x4.csv', skiprows=23) # 32 portfolios ME/BEME/OP

# Loading factor csv
fact_ff5 = ff_monthly_loader(f'{PATH}/F-F_research_Data_5_Factors_2x3.csv', skiprows=3) # 5 factor
```

```python
# Saving cleaned dataframes
fact_ff5.to_parquet(
    '../data/processed/ff5_factors_monthly.parquet',
    index=False
)

# 25 ME/SIZE
port_25_me_size.to_parquet(
    '../data/processed/ff_portfolios_25_ME_SIZE.parquet',
    index=False
)

# 25 ME/OP
port_25_me_op.to_parquet(
    '../data/processed/ff_portfolios_25_ME_OP.parquet',
    index=False
)

# 25 ME/INV
port_25_me_inv.to_parquet(
    '../data/processed/ff_portfolios_25_ME_INV.parquet',
    index=False
)

# 32 ME/BEME/INV
port_32_me_beme_inv.to_parquet(
    '../data/processed/ff_portfolios_32_ME_BEME_INV.parquet',
    index=False
)

# 32 ME/BEME/OP
port_32_me_beme_op.to_parquet(
    '../data/processed/ff_portfolios_32_ME_BEME_OP.parquet',
    index=False
)
```
